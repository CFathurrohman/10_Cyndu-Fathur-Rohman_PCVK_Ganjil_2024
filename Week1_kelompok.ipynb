1. Akses file image lokal (KTP/SIM) dan tampilkan menggunakan OpenCV dan Matplotlib

```python
import cv2 as cv
from google.colab import files
from google.colab.patches import cv2_imshow
import matplotlib.pylab as plt
```

```python
# Upload gammbar dari local
image_uploaded = files.upload()
```

```python
image_path = list(image_uploaded.keys())[0]
image = cv.imread(image_path)

# Menampilkan gambar menggunakan OpenCV
cv2_imshow(image)

# Menampilkan gambar menggunakan Matplotlib
plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
plt.axis('off')
```

2. Identifikasi kenapa menggunakan konversi warna BGR2RGB pada matplotlib dan tanpa
konversi pada opencv <br><br>
OpenCV Membaca gambar dalam format BGR sedangkan Matplotlib menampilkan gambar dalam format RGB, oleh karena itu konversi format dari BGR ke RGB dibutuhkan saat menampilkan gambar dari Matplotlib agar warna gambar ditampilkan dengan benar.

3. Tutup tiap bagian tertentu dari kartu identitas tersebut. (Kelompok 4 tutup bagian Alamat dan Kecamatan.)

```python
# Blok Alamat
atas_kiri_x = 190
atas_kiri_y = 185

tinggi_blok = 25
lebar_blok = 200

for y in range (atas_kiri_y, atas_kiri_y + tinggi_blok):
  for x in range (atas_kiri_x, atas_kiri_x + lebar_blok):
    image[y,x] = [255,0,0]

# Blok Kecamatan
atas_kiri_x = 190
atas_kiri_y = 255

tinggi_blok = 25
lebar_blok = 120

for y in range (atas_kiri_y, atas_kiri_y + tinggi_blok):
  for x in range (atas_kiri_x, atas_kiri_x + lebar_blok):
    image[y,x] = [0,0,255]

cv2_imshow(image)
```
