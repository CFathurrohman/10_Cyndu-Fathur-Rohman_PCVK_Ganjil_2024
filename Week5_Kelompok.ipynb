Gunakan image masking untuk mengambil bagian tertentu dari kartu identitas menggunakan
fungsi yang telah Anda pelajari. Kemudian masing-masing simpan dalam nama image baru<br><br>
Kelompok 10 ambil bagian RT/RW dan Kel/desa.  

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files  # Hanya diperlukan jika Anda menggunakan Google Colab

# Meng-upload gambar
uploaded = files.upload()  # Dialog untuk meng-upload file

# Mengambil nama file dari uploaded
image_path = list(uploaded.keys())[0]  # Ambil nama file dari dictionary

# Membaca gambar dengan OpenCV
image = cv.imread(image_path)
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)
```

```python
# Pertama: Masking untuk rt/rw
mask1 = np.zeros(image.shape[:2], dtype=np.uint8)
top_left1 = (60, 210)
bottom_right1 = (280, 230)
cv.rectangle(mask1, top_left1, bottom_right1, (255), thickness=-1)

# Mengaplikasikan mask ke gambar
masked_image1 = cv.bitwise_and(image_rgb, image_rgb, mask=mask1)

# Membuat gambar hasil masking dengan latar belakang putih
result_image1 = np.ones_like(image_rgb) * 255
result_image1[mask1 == 255] = masked_image1[mask1 == 255]

# Menyimpan hasil gambar pertama
cv.imwrite('hasil_masking_rt_rw.jpg', cv.cvtColor(result_image1, cv.COLOR_RGB2BGR))

# Kedua: Masking untuk kel/desa
mask2 = np.zeros(image.shape[:2], dtype=np.uint8)
top_left2 = (60, 233)
bottom_right2 = (250, 253)  # Sesuaikan agar tidak tumpang tindih
cv.rectangle(mask2, top_left2, bottom_right2, (255), thickness=-1)

# Mengaplikasikan mask ke gambar
masked_image2 = cv.bitwise_and(image_rgb, image_rgb, mask=mask2)

# Membuat gambar hasil masking dengan latar belakang putih
result_image2 = np.ones_like(image_rgb) * 255
result_image2[mask2 == 255] = masked_image2[mask2 == 255]

# Menyimpan hasil gambar kedua
cv.imwrite('hasil_masking_kel_desa.jpg', cv.cvtColor(result_image2, cv.COLOR_RGB2BGR))

# Tampilkan gambar asli dan hasil masking
plt.figure(figsize=(30, 15))

# Gambar asli
plt.subplot(1, 3, 1)
plt.imshow(image_rgb)
plt.title('Gambar Asli')

# Gambar hasil masking untuk rt/rw
plt.subplot(1, 3, 2)
plt.imshow(result_image1)
plt.title('Hasil Masking RT/RW')
plt.axis('off')

# Gambar hasil masking untuk Kel/Desa
plt.subplot(1, 3, 3)
plt.imshow(result_image2)
plt.title('Hasil Masking Kel/Desa')
plt.axis('off')

# Unduh file hasil masking
files.download('hasil_masking_rt_rw.jpg')
files.download('hasil_masking_kel_desa.jpg')
```

Hasil analisis : Kode diatas akan melakukan masking sebanyak dua kali pada gambar untuk mengekstrak bagian RT/RW dan Kel/Desa dari kartu identitas dimana bagian RT/RW dan Kel/Desa memiliki koordinat yang berbeda, Setelah dilakukan masking program akan menyimpan hasilnya sebagai hasil_masking_rt_rw.jpg dan hasil_masking_kel_desa.jpg, serta menampilkan gambar asli dan hasil masking secara berdampingan sebelum mengunduh kedua file tersebut.
