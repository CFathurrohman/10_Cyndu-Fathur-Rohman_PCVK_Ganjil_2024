Kelompok 4 tutup bagian Alamat dan Kecamatan.  

```python
import cv2 as cv
from google.colab import files
from google.colab.patches import cv2_imshow
import matplotlib.pylab as plt
```

```python
# Upload gammbar dari local
image_uploaded = files.upload()
```

```python
image_path = list(image_uploaded.keys())[0]
image = cv.imread(image_path)

# persegi panjang untuk menutupi alamat
cv.rectangle(image,pt1=(190,190),pt2=(400,210),color=(255,0,0),thickness=-1)

# looping circle untuk menutupi kecamatan
for i in range(7) :
  cv.circle(image, center=(200+i*15,266), radius=10, color=(0,0,255), thickness=-1)
# Menampilkan gambar menggunakan Matplotlib
plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
```
